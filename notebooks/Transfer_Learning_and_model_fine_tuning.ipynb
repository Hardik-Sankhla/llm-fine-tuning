## Section: Transfer Learning & Model Fine-Tuning

**Question: What is Transfer Learning and Fine-Tuning? Define with examples.**
- Transfer Learning is a machine learning approach where a model trained on one task (usually with a large dataset) is reused for a different but related task.
- Instead of training from scratch, we use learned features from a pre-trained model.
- It saves time, data, and computation cost.

**Example:**  
Using a pre-trained **ResNet** model (trained on ImageNet) to classify medical X-ray images (e.g., normal vs pneumonia).

---

### Fine-Tuning
- Fine-Tuning is a specific type of transfer learning where we further train part or all of the pre-trained model on the new dataset.
- Usually, early layers are kept fixed (or partially fixed), and later layers are updated.
- Helps the model adapt better to the target task.

**Example:**  
Take **BERT** (pre-trained on large text corpora) and fine-tune it on a sentiment dataset to classify reviews as positive/negative.

---

### Quick Difference

| Aspect | Transfer Learning | Fine-Tuning |
|---|---|---|
| Basic idea | Reuse pre-trained model features | Continue training pre-trained model |
| Training effort | Lower | Higher than basic transfer learning |
| Best for | Small datasets, fast setup | Better task-specific performance |

--- 

In short, **fine-tuning is an advanced step within transfer learning** to improve performance on a specific target task.


# Transfer Learning and Fine-Tuning with Convolutional Neural Networks (CNNs)

### Dataset:  ILSVRC2012 (ImageNet Large Scale Visual Recognition Challenge 2012) is a popular dataset for image classification tasks. It contains over 1.2 million images across 1000 classes.



In **Convolutional Neural Network (CNN)** architectures, common models include:

- **VGG16**: A deep CNN with 16 layers, known for its simplicity and effectiveness in image classification.
- **ResNet50**: A 50-layer CNN that uses residual connections to enable training of deeper networks.
- **InceptionV3**: A CNN architecture that uses inception modules to capture multi-scale features while keeping computation efficient.
- **MobileNetV2**: A lightweight architecture designed for mobile and embedded devices, using depthwise separable convolutions to reduce parameters and computational cost.
- **DenseNet121**: A CNN in which each layer is connected to every other layer in a feed-forward manner, improving feature reuse and reducing the number of parameters.
- **EfficientNetB0**: A model that scales depth, width, and resolution in a balanced way to achieve high accuracy with fewer parameters and lower computational cost.
- **Xception**: An architecture based on depthwise separable convolutions, designed to improve performance on image classification tasks.
- **NASNetMobile**: A CNN architecture developed using Neural Architecture Search (NAS), optimized for both accuracy and efficiency on mobile devices.

### Fine-Tuning Approach for CNNs

- Replace the final classification layer of the pre-trained CNN with a new output layer suitable for the target task.
- Freeze the weights of the earlier layers so that only the new classification layers are trained initially.
- Train the modified model on the new dataset for a few epochs using a lower learning rate.
- Optionally unfreeze some of the deeper layers and continue fine-tuning on the target dataset for improved performance.
